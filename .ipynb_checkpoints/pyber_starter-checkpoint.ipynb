```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "city_data.csv"
ride_data_to_load = "ride_data.csv"

# Read the City and Ride Data
city_df = pd.read_csv(city_data_to_load)
ride_df = pd.read_csv(ride_data_to_load)

#Combine the data into a single dataset
data_set = pd.merge(ride_df, city_df, on="city", how="outer")
data_set.head()
```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types

urban = pd.DataFrame(data_set.loc[(data_set["type"] == "Urban")])
suburban = pd.DataFrame(data_set.loc[(data_set["type"] == "Suburban")])
rural = pd.DataFrame(data_set.loc[(data_set["type"] == "Rural")])

# Average Fare ($), Total Number of Rides and Total Number of Drivers for Urbs cities
avg_fare_urban = urban.groupby(["city"])["fare"].mean()
total_rides_urban = urban.groupby(["city"])["ride_id"].count()
total_drivers_urban = urban.groupby(["city"])["driver_count"].value_counts

avg_fare_suburban = suburban.groupby(["city"])["fare"].mean()
total_rides_suburban = suburban.groupby(["city"])["ride_id"].count()
total_drivers_suburban = suburban.groupby(["city"])["driver_count"].value_counts

avg_fare_rural = rural.groupby(["city"])["fare"].mean()
total_rides_rural = rural.groupby(["city"])["ride_id"].count()
total_drivers_rural = rural.groupby(["city"])["driver_count"].value_counts

# Build the scatter plots for each city types
plt.scatter(total_rides_urban, avg_fare_urban, s=total_rides_urban*10,
             marker ='o', facecolors ="lightcoral", edgecolors='black',alpha = 0.75, label="Urban")
plt.scatter(total_rides_suburban, avg_fare_suburban, s=total_rides_suburban*10,
             marker ='o', facecolors ="lightblue", edgecolors='black',alpha = 0.75, label="Suburban")
plt.scatter(total_rides_rural, avg_fare_rural, s=total_rides_rural*10,
             marker ='o', facecolors ="lightgreen", edgecolors='black',alpha = 0.75, label="Rural")
plt.title("Pyber Ride Sharing Data (2016)", fontsize= "15", fontweight="bold")
plt.legend(loc='upper right', shadow=True)
plt.xlabel("Total Numbers of Rides Per City")
plt.ylabel("Average Fare($)")
plt.text(43, 33, "Note: Circle size correlates with driver count per city", fontweight="bold")

plt.savefig("Pyber Ride Sharing Scatter Plot.png")
```

## Observation No.1
#### Fares are more expensive in rural areas than they are in suburban or urban areas.

## Total Fares by City Type

```python
# Calculate Type Percents
num_fares = data_set.groupby(["type"])["fare"].sum()
colors = ["gold", "lightblue", "lightcoral"]
labels = ["Rural", "Suburban", "Urban"]
explode = (0, 0, 0.1)

# Build Pie Chart
plt.pie(num_fares, labels=labels, colors=colors, explode=explode, autopct="%1.1f%%", shadow=True, startangle=140)
plt.title("Total Fares by City Type", fontweight="bold", fontsize="13")

# Save Figure
plt.savefig("Total Fares by City Type.png")
plt.show()
```

## Total Rides by City Type

```python
# Calculate Rides Percents
num_rides = data_set.groupby(["type"])["ride_id"].count()
colors = ["gold", "lightblue", "lightcoral"]
labels = ["Rural", "Suburban", "Urban"]
explode = (0, 0, 0.1)

# Build Pie Chart
plt.pie(num_rides, labels=labels, colors=colors, explode=explode, autopct="%1.1f%%", shadow=True, startangle=140)
plt.title("Total Rides by City Type", fontweight="bold", fontsize="13")

# Save Figure
plt.savefig("Total Rides by City Type.png")
plt.show()
```

## Observation No.2
#### Most of uber customers live in urban areas.

## Total Drivers by City Type

```python
# Calculate Driver Percents
data_set_city = data_set.drop_duplicates('city')
num_drivers = data_set_city.groupby(["type"])["driver_count"].sum()
colors = ["gold", "lightblue", "lightcoral"]
labels = ["Rural", "Suburban", "Urban"]
explode = (0, 0, 0.1)

# Build Pie Chart
plt.pie(num_drivers, labels=labels, colors=colors, explode=explode, autopct="%1.1f%%", shadow=True, startangle=140)
plt.title("Total Drivers by City Type", fontweight="bold", fontsize="13")

# Save Figure
plt.savefig("Total Drivers by City Type.png")
plt.show()

```

## Observation No.3
#### A significant number of uber drivers work in urban areas and very few drivers drive in rural areas.
#### Urbanites are the ones that use uber services the most.
